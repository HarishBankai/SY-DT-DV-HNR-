```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/heart.csv')
print(df)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
print(df.dtypes)
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df)
plt.xlabel('trestbps')
plt.ylabel('chol')
plt.title('Boxplot : trestbps VS chol')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True, cmap='coolwarm')
plt.title('Correlation Matrix : ')
plt.show()
```

```python
print(df.corr())
```

```python
if (df.corr() == 0).any().any():
  print('Patient is healthy')
else:
  print('Patient is not healthy')
```
